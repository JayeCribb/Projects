## Baysian Hypothesis Testing

The aim of the following is to replace the frequentist t-test approach to hypothsis testing with its Baysian counterpart, using the Python library Pymc3. A Baysian posterior estimation will be used to look at the differneces between groups, replacing the t-distribution with the normal. This approach is often more informative than the frequentist approach due to the inherent updating process, and often results obtained are more readily interpretable.

The dataset being used contains records of different player statistics while they are playing a computer game called Starcraft. Baysian statistics will be used to look at the differences in APM (Actions Per Minute) between two different groups of players.

```python
# Load libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import pymc3 as pm
from scipy import stats
```

```python
# Load data

star = pd.read_csv('./assets/datasets/skillcraft1_dataset.csv')
star.head()
```

### Is there a significant difference in APM between 26 and 19 year old players?

```python
# Seperate APM into the desired groups of players
apm19 = star[star.Age == '19']["APM"].values
apm26 = star[star.Age == '26']["APM"].values
```

#### Setting up prior beliefs of the groups

```python
# Find the mean and standard deviation of APM regardless of age group, using them to inform the prior belief
mean_prior_mean = star["APM"].mean()
mean_prior_std = star["APM"].std()
```

```python
# Set up prior belief for the mean actions per minute for the two groups. This will be a normal distribution
# covering a range of possible values with varying liklihoods, with the highest liklihood at the mean. The 
# distributions for the two groups will be the same, as the prior belief is that the mean actions per minute are the
# same.

model=pm.Model()

with model:
    group19_mean = pm.Normal('APM_19yo_mean', mu = mean_prior_mean, sd = mean_prior_std)
    group26_mean = pm.Normal('APM_26yo_mean', mu = mean_prior_mean, sd = mean_prior_std)
```

```python
# Plot prior distributions

samples = [group19_mean.random()[0] for i in range(1000)]
plt.hist(samples, bins = 50)
plt.title('Mean Prior Distributions')
plt.show()
```

```python
# Set up prior belief for the standard deviation. The standard deviation priors will be uniformally distributed,
# inferring that all standard deviations within a range are equally likely.

# Standard deviation range
std_prior_lower = 0.01
std_prior_upper = 100.0

with model:
    group19_std = pm.Uniform('APM_19yo_std', lower = std_prior_lower, upper = std_prior_upper)
    group26_std = pm.Uniform('APM_26yo_std', lower = std_prior_lower, upper = std_prior_upper)
```

```python
# Plot prior distributions

samples = [group19_std.random()[0] for i in range(1000)]
plt.hist(samples, bins = 50)
plt.title('SD Prior Distributions')
plt.show()
```

#### Setting up Bayes

```python
# The observed data points are passed to the liklihood function, which we combine with the priors to return the
# posterior probability distribution for a given hypothesis. When the model is fitted using Markov Chain Monte Carlo,
# a random sampling process, the two posteria distributions will be updated according to the prior belief.

with model:
    group19 = pm.Normal('APM_19yo', mu = group19_mean, sd = group19_std, observed = apm19)
    group26 = pm.Normal('APM_26yo', mu = group26_mean, sd = group26_std, observed = apm26)
```

```python
# Setting up additional metrics, the ones we are interested in i.e differences between the two groups. Here
# we use Deterministic distributions, which are distibutions constructed from other distributions.
# Effect size is a normalised measure of the strength of a difference between the means.

with model:
    diff_of_means = pm.Deterministic('difference of means', group19_mean - group26_mean)
    diff_of_stds = pm.Deterministic('difference of stds', group19_std - group26_std)
    effect_size = pm.Deterministic('effect size', diff_of_means / np.sqrt((group19_std**2 + group26_std**2) / 2))
```

#### Fitting the model with  Markov Chain Monte Carlo (MCMC)

MCMC is a sampling algorithm that lets us construct arbitrary posteria distributions given the data and priors. The algorithm samples across different means and standard deviations (our hypothsis) and tests how likely each individual data point in the dataset is given the hypothesis. The posterior is then constructed by the product of all those likelihoods for each data point in the dataset for that one given hypothesis. This is computed for each mean and standard deviation in the trace set, dependant on the relative liklihoods, such that more samples are taken from higher likelihood regions of the hypothesis space. 

```python
# Fit model with MCMC
with model:
    trace = pm.sample(25000, njobs = 4)

burned_trace = trace[10000:] # Good practice to discard the first lot of traces due to convergence of liklihood
```

#### Plot Posteriors distributions

```python
# The HPD is the Highest Posterior Density interval. This gives us a bayesian "credible interval" which is the
# corollary to the frequentist "confidence interval". The 95% HPD says that the 95% highest density points on 
# the distribution fall within that range.

pm.plot_posterior(burned_trace, varnames = ['APM_19yo_mean', 'APM_26yo_mean', 'APM_19yo_std', 'APM_26yo_std'])
plt.show()
```

```python
pm.plot_posterior(burned_trace, varnames=['difference of means', 'difference of stds', 'effect size'], ref_val = 0)
plt.show()
```

What the above distributions show is that at a 95% significance level, there is a significant difference in means between the two groups, and also effect size. There is not a significant difference between the standard deviations of the two groups.

#### Frequentist approach: t-test with normal distributions

```python
# This is a two-sided test for the null hypothesis that 2 independent sample, have identical average (expected)
# values.
stats.ttest_ind(apm19, apm26, equal_var = False)
```
