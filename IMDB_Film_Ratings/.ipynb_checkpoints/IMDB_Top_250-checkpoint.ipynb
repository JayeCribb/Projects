## Predicting IMDB Film Ratings

The aim of the following is to extract features on IMDB top 250 films in order to build a model that accurately predicts film ratings. Data on films will be aquired by a combination of Web Scraping, using the Python library Beautiful Soup, and using an API. The work flow of the project will be aquiring the data, cleaning the data, exploratory data analysis and finally modelling.

```python
# Import Libraries

import pandas as pd
import numpy as np
import json
import requests
import re
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from urllib2 import urlopen
from bs4 import BeautifulSoup
import os
import subprocess
import scipy
import requests
import nltk
from nltk.corpus import stopwords
import collections
import csv
import psycopg2
from sklearn.preprocessing import Imputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.feature_extraction.text import CountVectorizer
```

#### Aquiring Data

```python
# URL of IMDB top 250 films
url = 'http://www.imdb.com/chart/top'

# Read IMDB page using BeautifulSoup
source = urlopen(url).read()
soup = BeautifulSoup(source, "lxml" )
```

```python
# Aquire film title IDs contained within the href
soup.find('td', class_='titleColumn').find('a', href = True)['href'][7:16]
```

```python
tds = soup.findAll('td', class_='titleColumn')
title_ids = []
for td in tds:
    for a in td.findAll('a', href=True):
        title_ids.append(a['href'][7:16])
```

```python
# Scrape reviews using film titles

reviews = []
for i in title_ids:
    address = 'http://www.imdb.com/title/' + i + '/reviews'
    html = requests.get(address).text
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.findAll('title')
    page = soup.findAll('p')
    review = str(page)
    regex = re.compile('[^a-zA-Z]')
    cleaned_review = regex.sub(' ', review)
    reviews.append([i,cleaned_review])
```

```python
# Create dataframe from reviews

review_df = pd.DataFrame(reviews, columns = ['id', 'review'])
```

```python
len(review_df)
```

```python
# Apply CountVectorizor to reviews

cv = CountVectorizer(stop_words = 'english', max_features = 100, ngram_range = (1, 2))
review_df = cv.fit_transform(review_df['review'])
```

```python
review_df = pd.DataFrame(review_df.toarray(), columns = cv.get_feature_names())
review_df.head()
```

```python
# Remove all but the top four descriptive words

review_df = review_df[['best', 'good', 'great', 'love']]
review_df.head(2)
```

```python
# Function that retrieves information on films from an API using movie id

def get_entry(entry):
    res = requests.get('http://www.omdbapi.com/?i='+entry)
    if res.status_code != 200:
        print entry, res.status_code
    else:
        print '.',
    try:
        j = json.loads(res.text)
    except ValueError:
        j = None
    return j
```

```python
# Get data

data = []
for id_ in title_ids:
    data.append(get_entry(id_))
```

```python
description = pd.DataFrame(data)
description.head(2)
```

```python
# Function that scrapes imdb for the gross revenue

def get_gross(entry):
    response = requests.get('http://www.imdb.com/title/'+ entry)
    html = response.text
    try:
        gross_list = re.findall("Gross:</h4>[ ]*\$([^ ]*)", html)
        gross = int(gross_list[0].replace(',', ''))
        print '.',
        return gross
    except Exception as ex:
        print ex, entry, response.status_code
        return None
```

```python
# Get data

gross = []
for id_ in title_ids:
    gross.append(get_gross(id_))
```

```python
# Create dataframe from revenue

revenue = pd.DataFrame(gross, columns = ['revenue'])
```

```python
# Combine dataframes

df = pd.concat([description, revenue, review_df], axis = 1)
df.head(2)
```

### Data Cleaning

```python
# Drop redundant columns

df = df.drop(['Language','Writer', 'Metascore', 'Plot', 'Poster', 'Rated', 'Response', 'Released', 'imdbID', 'Type'], axis = 1)
```

```python
df.info()
```

```python
# Convert data types

df['Year'] = df['Year'].astype('int64')
df['imdbRating'] = df['imdbRating'].astype('float64')
df['imdbVotes'] = map(lambda x: int(x.replace(',','')), df['imdbVotes'])
```

```python
## Combined number of awards and nominations from text

def get_num(text):
    try:
        result = sum([int(s) for s in text.split() if s.isdigit()])
    except:
        result = 0
    return result
```

```python
# Apply get_num function to Awards column

df['Awards / Noms'] = map(lambda x: get_num(x), df['Awards'])
df = df.drop(['Awards'], axis = 1)
```

```python
# Strip Runtime of text and convert to integer

df['Runtime'] = map(lambda x: x.replace('min', '').strip(), df['Runtime'])
df['Runtime'] = df['Runtime'].astype('int64')
```

```python
# Impute missing revenue values with the mean of the column

im = Imputer()
df['revenue'] = im.fit_transform(df['revenue'].values.reshape(-1, 1))
```

### Modelling

```python
# Apply Countvectoriser to desired columns

cv = CountVectorizer(tokenizer=lambda x: x.split(', '), max_features = 10)
countries = cv.fit_transform(df['Country']).todense()
countries = pd.DataFrame(countries, columns = cv.get_feature_names())
```

```python
cv = CountVectorizer(tokenizer=lambda x: x.split(', '), max_features = 10)
actors = cv.fit_transform(df['Actors']).todense()
actors = pd.DataFrame(actors, columns = cv.get_feature_names())
```

```python
cv = CountVectorizer(tokenizer=lambda x: x.split(', '), max_features = 10)
directors = cv.fit_transform(df['Director']).todense()
directors = pd.DataFrame(directors, columns = cv.get_feature_names())
```

```python
cv = CountVectorizer(tokenizer=lambda x: x.split(', '), max_features = 10)
genres = cv.fit_transform(df['Genre']).todense()
genres = pd.DataFrame(genres, columns = cv.get_feature_names())
```

```python
# Create dataframe

df = pd.concat([df, countries, actors, directors, genres], axis = 1)
```

```python
# Drop redundant columns

df = df.drop(['Actors', 'Director', 'Genre', 'Country'], axis = 1)
```

```python
# Define feature and target set

X = df.drop(['Title', 'imdbRating' ], axis = 1)
y = df['imdbRating']
```

```python
# Train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Function to calculate metrics and plot scatters for training and test set

def plot_scatter(X_train, X_test, y_train, y_test, model):
    predictions_train = model.predict(X_train)
    predictions_test = model.predict(X_test)
    print 'Train MSE: ', mean_squared_error(y_train, predictions_train)
    print 'Train R2: ', r2_score(y_train, predictions_train)
    print '----------------------------------------------------------------------------------'
    print 'Test MSE: ', mean_squared_error(y_test, predictions_test)
    print 'Test R2: ', r2_score(y_test, predictions_test)
    print '----------------------------------------------------------------------------------'
    
    plt.figure(figsize = (16, 7))
    
    plt.subplot(1,2,1)
    plt.title('Training Set Predictions', fontsize = 15)
    plt.scatter(y_train, predictions_train)
    plt.plot(y_train, y_train, c = 'r')
    plt.xlabel('Actual value')
    plt.ylabel('Predicted value')
    plt.subplot(1,2,2)
    plt.title('Test Set Predictions', fontsize = 15)
    plt.scatter(y_test, predictions_test)
    plt.plot(y_test, y_test, c = 'r')
    plt.xlabel('Actual value')
    plt.ylabel('Predicted value')
    
    plt.show()
```

```python
# Random Forest Regressor

rf = RandomForestRegressor()
model = rf.fit(X_train, y_train)

plot_scatter(X_train, X_test, y_train, y_test, model)
```

```python
# Optimised Random Forests

rf = RandomForestRegressor()
param_grid = {'n_estimators': np.arange(50, 301, 50),
             'max_depth': [None, 2, 4, 6],
             'min_samples_split': [2, 4, 6]}
model = GridSearchCV(rf, param_grid = param_grid, cv = 5, verbose = 1)
fitted_model = model.fit(X_train, y_train)

plot_scatter(X_train, X_test, y_train, y_test, fitted_model)
```

```python
print 'Optimised Parmeters: ', fitted_model.best_params_
```

```python
# Feature importance

rf = RandomForestRegressor(max_depth = None, min_samples_split = 6, n_estimators = 100)
model = rf.fit(X_train, y_train)

importances = rf.feature_importances_ # Feature importance
std = np.std([tree.feature_importances_ for tree in rf.estimators_], axis = 0) # Standard deviation of importance 
indices = np.argsort(importances)[::-1] # Sort importance in descending order

# Plot the feature importances of the forest
plt.figure(figsize = (16, 6))
plt.title("Feature importances", fontsize = 15)
plt.bar(range(X_train.shape[1]), importances[indices], yerr = std[indices], align="center", color = 'gray')
plt.xticks(range(X_train.shape[1]), X_train.columns[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.show()
```

The most important feature when modelling IMDB ratings is the number of votes the movie recieves. This feature dominates all others and in general the more votes the movie recieves, the higher the rating. Other features important to the modelling process includes the year in which the film was made, the number of times the word 'good' appeared in the review and the number of awards/nominations the movie recieved.
