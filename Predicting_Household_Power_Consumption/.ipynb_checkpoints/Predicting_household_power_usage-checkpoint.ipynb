## Predicting Household Power Usage

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('household_power_consumption.txt', delimiter = ';')
```

```python
df.head(3)
```

```python
# Power usage per minute 

power = df['Global_active_power']
```

```python
# Convert power usage to float. In the case of special characters, pass.

converted_power = []
for i in power:
    try:
        converted_power.append(float(i))
    except ValueError:
                pass
```

```python
# Plot 24hr period of usage

plt.figure(figsize = (16, 6))
plt.plot(converted_power[: 24*60])
plt.show()
```

```python
# Scale data using Standard scaler so that it has 0 mean and unit variance

st = StandardScaler()
scaled_power = st.fit_transform(pd.DataFrame(converted_power))
```

```python
reduced_data = scaled_power[:1000]
```

```python
# Create a sequence dataframe using data from the previous 59 minutes, which will be the input to the LSTM.
# The target that we want to predict is time step t+1 using the features at time steps t-59 through to t.

y = [reduced_data[i + 60, 0] for i in range(len(reduced_data) - 60 - 1)]
X = [reduced_data[i:(i + 60), 0] for i in range(len(reduced_data) - 60 - 1)]
X_df = pd.DataFrame(X)
X_df.head(2)
```

```python
# Create a train test split that respects order

X_train = np.array(X_df[0: int(len(X_df) * 0.7)])
X_test = np.array(X_df[int(len(X_df) * 0.7):])
y_train = np.array(y[0: int(len(X_df) * 0.7)])
y_test = np.array(y[int(len(X_df) * 0.7):])
```

```python
# Reshape data in the form [samples, timesteps, features] such that the sequence of 59 timesteps is one feature.

X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0],  X_test.shape[1], 1)
```

```python
# Define model

model = Sequential()

model.add(LSTM(50, input_dim = 1, return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(100, input_dim = 1, return_sequences = False))
model.add(Dropout(0.2))
model.add(Dense(1, activation = 'linear'))

model.compile(loss = "mse", optimizer = "adam")
```

```python
model.fit(X_train, y_train, nb_epoch = 20, batch_size = 1, verbose = 1)
```

```python
test_predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test, test_predictions)
```

```python
plt.figure(figsize = (16, 6))
plt.plot(test_predictions, label = 'Prediction')
plt.plot(y_test, label = 'True Outcome')
plt.legend()
plt.show()
```
