## Baysian Regression

The aim of the following is to apply Baysian statistics to model data using Linear Regression. When Regression is performed using Baysian Statistics a posterior distribution is created, which enables us to see the entire range of values for the intercept and coefficients of the model, along with the degree of uncertainty in the estimates. Furthermore, knowledge about the prior can be included in the modelling process.

```python
# Load libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import pymc3 as pm
from sklearn.linear_model import LinearRegression
from sklearn import datasets
%matplotlib inline
```

```python
# Generate line of best fit along with random noise to create points

intercept = 4
slope = 3
x = np.linspace(0, 1, 200)
true_regression_line = intercept + slope * x

# Create points by generating error terms
y = true_regression_line + np.random.normal(scale = 0.25, size = 200)
# Create dataframe to use with pymc3
data = pd.DataFrame(data = dict(x = x, y = y))
```

```python
# Plot scatter with 'true' regression line

plt.figure(figsize = (10, 6))
plt.scatter(x, y, s = 10)
plt.plot(x, true_regression_line, color = 'r', label = 'Regression Line')
plt.legend(loc = 2, fontsize = 15)
plt.xlim((0,1))
plt.show()
```

```python
# Baysian Linear Regression

# Define generalised linear model
linear_model = pm.Model() 

with linear_model:
    pm.glm.glm('y ~ x', data)
    trace = pm.sample(5000, njobs = 2) # MCMC sampling algorithm is NUTS 

# Burn the first 2000 trace samples from the total
burned_trace=trace[2000:]
```

```python
# Plot trace output

pm.traceplot(burned_trace)
plt.tight_layout()
plt.show()
```

The graphs on the left-hand side represents the posterior distributions for the intercept, the slope and the standard deviation of the noise. The posterior distributions were constructed by the random sampling method, which compares the liklihood of possible values from the hypothesis space given the data. The priors are assumed to be Normally/Gaussian distributed. The peaks in the data represent the most likely values for our parameters. As can be seen, they are very similar to the coefficients of the 'true' line of best fit (intercept = 4, slope = 3, sd = 0.25).

The graphs on right-hand side represents the sampling chain of hypothesis - this shows the stability of sampling

```python
# Plot regression lines using posterior distribution

plt.figure(figsize = (10, 6))
plt.scatter(x, y, s = 10)
pm.glm.plot_posterior_predictive(burned_trace, samples = 500, label = 'Posterior Regression Lines', c = 'lightblue')
plt.plot(x, true_regression_line, color = 'r', label = 'Regression Line')
plt.legend(loc = 2, fontsize = 15)
plt.xlim((0,1))
plt.show()
```

```python
# Summary to show the uncertainty of each parameter

pm.summary(burned_trace)
```

```python
# Comparison with Sklearn

lr = LinearRegression()
model = lr.fit(data['x'].values.reshape(-1,1), data['y'])
print 'Intercept: ',model.intercept_
print 'Slope: ', model.coef_[0]
```

#### Using Diabetes Dataset

```python
# Load the diabetes dataset
data, target = datasets.load_diabetes(return_X_y = True)
```

```python
# Scatter plot of third variable with target
plt.figure(figsize = (10, 6))
plt.scatter(data[:,2], target, s = 10)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.show()
```

```python
# Linear Regression using Sklearn

lr = LinearRegression()
model = lr.fit(data[:,2].reshape(-1,1), target)
print 'Intercept: ',model.intercept_
print 'Slope: ', model.coef_[0]
```

```python
# Plot regression line

Regression_line = model.intercept_ + model.coef_[0] * data[:,2]

plt.figure(figsize = (10, 6))
plt.plot(data[:,2], Regression_line, c = 'r')
plt.scatter(data[:,2], target, s = 10)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.show()
```

```python
# Create dataframe to use with pymc3

df = pd.DataFrame(data = dict(x = data[:,2] , y = target))
```

```python
# Baysian Linear Regression

# Define linear model
linear_model = pm.Model() 

with linear_model:
    pm.glm.glm('y ~ x', df)
    trace = pm.sample(5000, njobs = 2) # MCMC sampling algorithm is NUTS 

# Burn the first 2000 trace samples from the total
burned_trace=trace[2000:]
```

```python
# Plot trace output

pm.traceplot(burned_trace)
plt.tight_layout()
plt.show()
```

```python
# Plot regression lines using posterior distribution

plt.figure(figsize = (10, 6))
plt.scatter(data[:,2], target, s = 10)
pm.glm.plot_posterior_predictive(burned_trace, samples = 500, label = 'Posterior Regression Lines', c = 'lightblue')
plt.plot(data[:,2], Regression_line, c = 'r', label = 'Sklearn Regression Line')
plt.legend(loc = 2, fontsize = 15)
plt.xlim((-0.15, 0.2))
plt.ylim((0, 400))
plt.show()
```

```python
# Summary

pm.summary(burned_trace)
```

```python
# Manually selecting priors and liklihood distributions

model = pm.Model()

with model:
    
    intercept = pm.Normal('intercept', mu = 0, sd = 1000)
    coef = pm.Normal('coef', mu = 0, sd = 1000)
    target_est = intercept + coef * data[:,2]
    model_error = pm.Uniform('model_error', lower = 0, upper = data[:,2].std()*1000)
    likelihood = pm.Normal('target_like', mu = target_est, sd = model_error, observed = data[:,2])
```
