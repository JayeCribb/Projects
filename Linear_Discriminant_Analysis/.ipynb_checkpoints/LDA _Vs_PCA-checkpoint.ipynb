## Linear Descriminant Analysis Vs Principal Component Analysis

The aim of the following is to look at the differences bewteen Linear Descriminant Analysis and Principal Component Analysis. To do this I will be using housing data taken from Boston, and modelling prices as a classification problem using both the LDA and PCA transformed data.

```python
# Load libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import itertools
```

```python
# Load Boston data from Sklearn

boston = datasets.load_boston()
X = boston.data # Feature set
y = boston.target # Housing prices
feature_names = boston.feature_names 
```

```python
# Transform the continuous target variable into four different class labels according to which quartile 
# of y the given value belongs to.

# Define quartiles
lower = np.percentile(y, q = 25)
mid = np.percentile(y, q = 50)
upper = np.percentile(y, q = 75)

# Quartile label function

def percentile(y):
    if y <= lower:
        return 0
    elif y > lower and y <= mid:
        return 1
    elif y > mid and y <= upper:
        return 2
    else:
        return 3
```

```python
# Apply function to house prices

target = map(percentile, y)
```

```python
# Normalise the feature set, important for both PCA and LDA to ensure that features with a larger scale
# do not dominate.

sc = StandardScaler()
X = sc.fit_transform(X)
```

![](LDA_diagram.png)

**Transform data using PCA and LDA**

PCA and LDA are used to reduce the dimentionality of data to produce models which generalise better. PCA works by 
finding the eigen-vector / eigen-value decomposition of the covariance matrix for the feature set. Essentially it splits the feature space into a number of principal components, each split maximising the variance in the feature space. When maximising the variance, PCA does not take into account classification of labels, which could result in the mixing of classes.

When data is transformed using LDA, the new reduced feature set minimises the dispersion between samples of the same class, and maximises the dispertion between samples of different classes, maintaining the seperation
of classes.

```python
pca = PCA(n_components = 2)
pca_reduced = pca.fit_transform(X)
cum_exp_var_pca = np.cumsum(pca.explained_variance_ratio_)

lda = LDA(n_components = 2)
lda_reduced = lda.fit_transform(X, target)
cum_exp_var_lda = np.cumsum(lda.explained_variance_ratio_)

print 'PCA cumulative explained variance ratio: ', cum_exp_var_pca
print 'LDA cumulative explained variance ratio: ', cum_exp_var_lda
```

As shown above, because LDA respects the seperation of classes, the explained variance ratio is significantly better than the explained variance ratio for the PCA transformed dataset.

```python
# Plot PCA and LDA 

plt.figure(figsize = (18, 8))

plt.subplot(1, 2, 1)
plt.title('PCA Transformed Data')
plt.scatter(pca_reduced[:,0], pca_reduced[:,1], c = target, cmap = 'rainbow')
plt.xlabel('First_principal_component')
plt.ylabel('Second_principal_component')

plt.subplot(1, 2, 2)
plt.title('LDA Transformed Data')
plt.scatter(-lda_reduced[:,0], -lda_reduced[:,1], c = target, cmap = 'rainbow')
plt.xlabel('First_LDA_component')
plt.ylabel('Second_LDA_component')

plt.show()
```

As can be shown by the above plot, there is a lot less mixing of classes when transforming data using LDA compared to PCA.

```python
# Classification using LDA

X_train, X_test, y_train, y_test = train_test_split(lda_reduced, target, test_size = 0.3, random_state = 42)
model = lda.fit(X_train,y_train)
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```

#### Metrics

```python
# Confusion matrix function

def plot_confusion_matrix(cm, classes, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                    horizontalalignment="center",
                    color="white" if cm[i, j] > thresh else "black")
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    return
```
